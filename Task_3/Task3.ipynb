```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load CSV
try:
    df = pd.read_csv("student_info.csv")
    print(" Data loaded successfully!\n")
except Exception as e:
    print(f" Could not load file: {e}")
```

```python
# Show first few rows
print("\n--- First 5 rows ---\n")
print(df.head())
```

```python
# Show columns
print("\n--- Columns in dataset ---\n")
print(df.columns.tolist())
```

```python
# Try to use these expected columns if they exist:
expected_cols = ['Student_ID', 'Name', 'Gender', 'Age', 'Marks', 'Attendance']
```

```python
# Check and warn if missing
for col in expected_cols:
    if col not in df.columns:
        print(f"⚠️  Column '{col}' not found, will skip plots needing it.")
```

```python
# Plot 1: Marks Distribution
if 'Marks' in df.columns:
    plt.figure(figsize=(8,4))
    sns.histplot(df['Marks'], bins=20, kde=True)
    plt.title("Distribution of Marks")
    plt.xlabel("Marks")
    plt.ylabel("Number of Students")
    plt.show()
```

```python
# Plot 2: Attendance Distribution
if 'Attendance' in df.columns:
    plt.figure(figsize=(8,4))
    sns.histplot(df['Attendance'], bins=20, kde=True, color="orange")
    plt.title("Distribution of Attendance")
    plt.xlabel("Attendance (%)")
    plt.ylabel("Number of Students")
    plt.show()
```

```python
# Display the columns clearly
print("Columns in your dataset:")
for col in df.columns:
    print(col)
```

```python
# Plot 3: Marks by Gender
if 'Marks' in df.columns and 'Gender' in df.columns:
    plt.figure(figsize=(8,4))
    sns.boxplot(x="Gender", y="Marks", data=df, palette="pastel")
    plt.title("Marks by Gender")
    plt.ylabel("Marks")
    plt.show()
```

```python
# Plot 4: Attendance by Gender
if 'Attendance' in df.columns and 'Gender' in df.columns:
    plt.figure(figsize=(8,4))
    sns.boxplot(x="Gender", y="Attendance", data=df, palette="coolwarm")
    plt.title("Attendance by Gender")
    plt.ylabel("Attendance (%)")
    plt.show()
```

```python
# Plot 5: Age Distribution
if 'Age' in df.columns:
    plt.figure(figsize=(8,4))
    sns.histplot(df['Age'], bins=15, kde=True, color="green")
    plt.title("Age Distribution")
    plt.xlabel("Age")
    plt.ylabel("Number of Students")
    plt.show()
```

```python
# Plot 6: Correlation Heatmap for numeric columns
numeric_cols = df.select_dtypes(include='number').columns.tolist()
if len(numeric_cols) >= 2:
    plt.figure(figsize=(8,6))
    sns.heatmap(df[numeric_cols].corr(), annot=True, cmap="coolwarm")
    plt.title("Correlation Heatmap")
    plt.show()
```
