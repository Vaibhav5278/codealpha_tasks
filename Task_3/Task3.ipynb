```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load CSV
try:
    df = pd.read_csv("student_info.csv")
    print(" Data loaded successfully!\n")
except Exception as e:
    print(f" Could not load file: {e}")
```

```python
# Show first few rows
print("\n--- First 5 rows ---\n")
print(df.head())
```

```python
# Show columns
print("\n--- Columns in dataset ---\n")
print(df.columns.tolist())
```

```python
# 1️⃣ Bar Chart: Average math score by grade
avg_math = df.groupby("grade_level")["math_score"].mean()
avg_math.plot(kind="bar", color="skyblue")
plt.title("Average Math Score by Grade Level")
plt.xlabel("Grade Level")
plt.ylabel("Average Math Score")
plt.show()
```

```python

# 2️⃣ Histogram: Study hours
plt.hist(df["study_hours"], bins=10, color="green", edgecolor="black")
plt.title("Study Hours Distribution")
plt.xlabel("Study Hours")
plt.ylabel("Number of Students")
plt.show()

```

```python
# 3️⃣ Pie Chart: Gender distribution
gender_counts = df["gender"].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct="%.1f%%", startangle=90)
plt.title("Gender Distribution")
plt.show()
```

```python
# 4️⃣ Line Chart: Average attendance by grade
avg_attendance = df.groupby("grade_level")["attendance_rate"].mean()
avg_attendance.plot(kind="line", marker="o", color="purple")
plt.title("Average Attendance Rate by Grade")
plt.xlabel("Grade Level")
plt.ylabel("Attendance Rate (%)")
plt.grid(True)
plt.show()
```

```python
# 5️⃣ Box Plot: Writing scores by gender
sns.boxplot(x="gender", y="writing_score", data=df, palette="Set2")
plt.title("Writing Scores by Gender")
plt.ylabel("Writing Score")
plt.show()
```

```python

# 7️⃣ Countplot: Final results
plt.figure(figsize=(6,5))
sns.countplot(x="final_result", data=df, palette="pastel")
plt.title("Final Result (Pass/Fail)")
plt.ylabel("Number of Students")
plt.show()
```

```python
# 9️⃣ Bar chart: Parent education levels
plt.figure(figsize=(10,5))
sns.countplot(x="parent_education", data=df, order=df['parent_education'].value_counts().index, palette="cubehelix")
plt.title("Parent Education Level Distribution")
plt.xticks(rotation=45)
plt.ylabel("Count")
plt.show()
```
